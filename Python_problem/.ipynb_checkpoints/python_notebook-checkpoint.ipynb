# JSON to JSONL

```python
import json

profile = 'midasIIITD'

tweets = api.user_timeline(screen_name = profile,count=200)

JSON_file = [json.loads(json.dumps(t._json)) for t in tweets]


with open('output.jsonl', 'w') as outfile:
    for entry in JSON_file:
        json.dump(entry, outfile)
        outfile.write('\n')
```

# checking for images in tweets

```python
import requests

timeline = api.user_timeline(screen_name = profile,count=200)

for tweet in timeline: 
   for media in tweet.entities.get("media",[{}]):
      if media: 
            print(media['media_url_https'],'\n')
```

# Final

```python
with open('output.jsonl', 'r') as outfile:
    for entry in outfile:
        print('\n')
        entry = json.loads(entry)
        print('Tweet :',entry['text'])
        print('Created at :',entry['created_at'])
        print('Retweet Counts :',entry['retweet_count'])
        print('Tweet favorites :',entry['favorite_count'])
        try :print('Retweet favorites :',entry['retweeted_status']['favorite_count'])
        except:pass
        for media in entry['entities'].get("media",[{}]):
          if media: print('No of images :',len(media['media_url_https'].split(',')))
          else:print('No of images: None')
        
```
